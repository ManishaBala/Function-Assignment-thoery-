##1. What is the difference between a function and a method in Python?

-> Functions: A function is a structured, reusable code block utilized for carrying out a specific task. Functions facilitate the creation of modular code, simplifying maintenance and reusability. It is defined using the def keyword and can be called independently.


```python
#example for function

def greet(name):
    return f"Hello, {name}!"

print(greet("Manisha"))
```

-> Method: A method is similar to a function but is associated with an object.
It is called on an object (usually an instance of a class) and often works with the data (attributes) of that object. Methods are defined inside classes.

```python
#example for Method

class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Manisha"))
```

##5. What are iterators in Python and how do they differ from iterables?

-> Iterators: An iterator is an object that allows traversal through a countable number of values. It is used to iterate over iterable objects like lists, tuples, sets, etc

In Python, an iterator is an object that keeps track of the state of iteration and retrieves elements one at a time using __next__(). An iterable is any object that can return an iterator via the __iter__() method and can be iterated over multiple times. The key difference is that iterables are the source, while iterators perform the actual iteration.

```python
# example
numbers = [10, 20, 30]

# Create an iterator from the iterable
iterator = iter(numbers)

print(next(iterator))  # Manually accessing elements using next()
print(next(iterator))
print(next(iterator))

#here numbers is an iterable (a list).
#while iterator is an iterator created from the iterable, and we manually retrieve elements using next().

```

##6. Explain the concept of generators in Python and how they are defined.

-> A generator is a special type of iterator in Python that allows us to iterate over a sequence of values, but unlike regular iterators, generators generate values on the fly rather than storing them in memory all at once. This makes generators memory-efficient and ideal for working with large data sets.
A generator function using the def keyword, just like a normal function, but instead of return, we use yield to yield values.


Generators differ from regular functions and iterators in several ways, primarily in terms of their behavior, memory usage, and how they handle iteration.





```python
#Example

def generate_numbers():
    yield 1
    yield 2
    yield 3
    yield 4

gen = generate_numbers()

# Using next() to get values one by one
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# Here the generator generate_numbers() yields values one at a time, meaning it computes values lazily, consuming less memory.
```

##7. What are the advantages of using generators over regular functions?

-> Generators offer several key advantages over regular functions, especially in terms of memory efficiency, performance, and ease of use. Hereâ€™s a breakdown of the major benefits:

| **Advantage**                          | **Generators**                          | **Regular Functions**                 |
|----------------------------------------|-----------------------------------------|---------------------------------------|
| **Memory Usage**                       | Memory-efficient, produces one value at a time | Stores all values in memory           |
| **Computation**                        | Values computed on demand (lazy evaluation) | All values computed at once           |
| **Suitability for Large Data**         | Ideal for large datasets, reduces memory overhead | Not suitable for large datasets unless fully processed at once |
| **Code Simplicity**                    | Simpler to write, no need for explicit iterators | More boilerplate code required to handle iteration |
| **Handling Infinite Sequences**       | Can handle infinite sequences without memory overflow | Cannot handle infinite sequences efficiently |


```python
#example Generator that yields one value at a time

def generate_numbers(n):
    for i in range(n):
        yield i

# This only generates one number at a time
gen = generate_numbers(1000000)

# Print the first 5 values
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

##8. What is a lambda function in Python and when is it typically used?

-> A lambda function in Python is a small, anonymous function that is defined using the lambda keyword instead of the def keyword. Lambda functions can have any number of arguments but only one expression, which is evaluated and returned.

Lambda function is used when:

- we need a simple function for a short period and donâ€™t want to formally define it with def.

- When using functional programming constructs like map(), filter(), or reduce().

- When we need to pass a function as an argument to another function, like custom sorting or applying a function to elements.

```python
#example

square = lambda x: x*x
print(square(6))
```

##9. Explain the purpose and usage of the `map()` function in Python.

-> The map() function in Python is a built-in function that allows us to apply a given function to all items in an iterable (like a list, tuple, etc.) and returns an iterator that produces the results.

The map() function applies the provided function to each item in the iterable and returns a map object, which is an iterator. We can then convert the map object to a list, tuple, or any other iterable type to see the results.

```python
#example

# Regular function to square a number
def square(x):
    return x * x

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Applying the square function to each element in the list using map
squared_numbers = map(square, numbers)

# Converting the map object to a list and printing the result
print(list(squared_numbers))
```

##10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

-> map(): Transforms each element of an iterable independently and returns an iterator with the results.

filter(): Filters elements from an iterable based on a condition and returns an iterator with the elements that satisfy the condition.

reduce(): Reduces an iterable to a single value by applying a binary function cumulatively.

###When to Use Each Function:

Use map(): When we need to transform or apply a function to each element of an iterable individually (e.g., squaring a number, converting data types).

Use filter(): When we need to filter out elements from an iterable based on a condition (e.g., getting all even numbers, removing null values).

Use reduce(): When we need to reduce an iterable to a single cumulative value, such as computing the sum or product of numbers.

```python
#example1: map() (Converting Strings to Uppercase)

words = ["hello", "world", "python"]
uppercase_words = map(lambda x: x.upper(), words)
print(list(uppercase_words))
```

```python
#example 2: filter() (Selecting Numbers Greater than 10)

numbers = [5, 10, 15, 20, 25]
filtered_numbers = filter(lambda x: x > 10, numbers)
print(list(filtered_numbers))
```

```python
#example 3: reduce() (Calculating the Sum of Numbers)

from functools import reduce

numbers = [1, 2, 3, 4, 5]
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print(sum_of_numbers)
```
